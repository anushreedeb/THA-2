```
census_state = pd.read_excel("census_state.xlsx")

dhs_state = pd.read_csv("dhs_state.csv")

ihds_state = pd.read_csv("ihds_state.csv")


print(census_state.head())

print(dhs_state.head())

print (ihds_state.head())
```

```
census_state_col = census_state[["State"]]
print(census_state_col.sample(15))

ihds_state_col = ihds_state[["STATEID"]]
print(ihds_state_col.sample(15))

dhs_state_col = dhs_state[["hv024"]]
print(dhs_state_col.sample(15))
```

census_ihds = census_state_col.merge(ihds_state_col, left_on='State' , right_on='STATEID', how="outer")

```

census_state_fnl = pd.DataFrame(census_state_col.drop_duplicates()['State'].str.lower())
print(census_state_fnl)

dhs_state_fnl = pd.DataFrame(dhs_state_col.drop_duplicates()['hv024'].str.lower())
print(dhs_state_fnl)

ihds_state_fnl = pd.DataFrame(ihds_state_col.drop_duplicates()['STATEID'].str.lower())
print(ihds_state_fnl)
```

```
ihds_state_new = pd.DataFrame(ihds_state_fnl['STATEID'].apply(lambda x:x[:-3]))
print(ihds_state_new)
```

```
census_dhs = census_state_fnl.merge(dhs_state_fnl, left_on='State', right_on='hv024')
print(census_dhs)

census_dhs_ihds = census_dhs.merge(ihds_state_new, left_on='State' , right_on='STATEID')
print(census_dhs_ihds)
```
